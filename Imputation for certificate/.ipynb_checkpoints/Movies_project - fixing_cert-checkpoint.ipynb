```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the imbd dataset
movies_file = 'imdb_1000_genre.csv'
# read the CVS file of the movies
imdb_1000_df = pd.read_csv(movies_file)
```

```python
imdb_1000_df.shape
```

```python
# Showing the head of the dataset
imdb_1000_df.head()
```

```python
# This line is for finding out the missing values in the dataset
missing_values = imdb_1000_df.isna().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot heatmap to visualize missing values
plt.figure(figsize=(12, 8))
sns.heatmap(imdb_1000_df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in IMDb Top 1000 Movies Dataset')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for the correlation matrix
numeric_columns = imdb_1000_df[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']]

# Convert the 'Gross' column to string, handle NaN values, and then remove commas
imdb_1000_df['Gross'] = imdb_1000_df['Gross'].astype(str).str.replace(',', '')

# Convert the 'Gross' column back to numeric (float), handling errors in case of any non-convertible values
imdb_1000_df['Gross'] = pd.to_numeric(imdb_1000_df['Gross'], errors='coerce')

# Calculating the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of the numerical variables of the dataset")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a count plot to visualize the relationship between 'Genre_1' and 'Certificate'
plt.figure(figsize=(10, 6))
sns.countplot(data=imdb_1000_df, x='Genre_1', hue='Certificate')

# Adding title and labels
plt.title("Distribution of 'Genre_1' across different 'Certificate' categories")
plt.xlabel("Primary Genre")
plt.ylabel("Count")

# Rotating the x labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Replace 'Approved', 'Passed', and 'Unrated' in the 'Certificate' column with NaN
imdb_1000_df['Certificate'] = imdb_1000_df['Certificate'].replace(['Approved', 'Passed', 'Unrated'], np.nan)

# Saving the modified dataset to a new CSV file
filtered_file_path = 'imdb_filtered_genre.csv'
imdb_1000_df.to_csv(filtered_file_path, index=False)
```

```python
# This line is for finding out the missing values in the dataset
imdb_1000_filtered_df = pd.read_csv(filtered_file_path)
missing_values = imdb_1000_filtered_df.isna().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
imdb_1000_filtered_df = pd.read_csv(filtered_file_path)

# Define a dictionary to map old values to new ones
certificate_mapping = {
    'G': 'U',
    'U/A': 'UA',
    'GP': 'PG',
    'TV-PG': 'PG',
    'TV-MA': 'A',
    'TV-14': 'PG-13',
    '16': 'PG-13'
    # Add more mappings as needed
}

# Apply the mapping to the 'Certificate' column
imdb_1000_filtered_df['Certificate'] = imdb_1000_filtered_df['Certificate'].replace(certificate_mapping)

# Optionally, save the modified dataset to a new file
mapped_file_path = 'imdb_mapped_certificate.csv'
imdb_1000_filtered_df.to_csv(mapped_file_path, index=False)

print(f"Modified dataset saved to {mapped_file_path}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
imdb_1000_mappped_df = pd.read_csv(mapped_file_path)


# Creating a count plot to visualize the relationship between 'Genre_1' and 'Certificate'
plt.figure(figsize=(10, 6))
sns.countplot(data=imdb_1000_mappped_df, x='Genre_1', hue='Certificate')

# Adding title and labels
plt.title("Distribution of 'Genre_1' across different 'Certificate' categories")
plt.xlabel("Primary Genre")
plt.ylabel("Count")

# Rotating the x labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
meta_file_path = 'movie_metadata.csv'
imdb_meta_df = pd.read_csv(meta_file_path)

# Make sure the titles are stripped of leading/trailing spaces for consistency
imdb_meta_df['movie_title'] = imdb_meta_df['movie_title'].str.strip()
imdb_1000_mappped_df['Series_Title'] = imdb_1000_mappped_df['Series_Title'].str.strip()

# Perform a left merge to fill missing values in 'Certificate' from 'content_rating' where titles match
merged_data = pd.merge(imdb_1000_mappped_df, imdb_meta_df[['movie_title', 'content_rating']], 
                       left_on='Series_Title', right_on='movie_title', how='left')

# Fill missing values in 'Certificate' with the 'content_rating' from the second dataset
merged_data['Certificate'] = merged_data['Certificate'].fillna(merged_data['content_rating'])

# Drop the extra 'movie_title' and 'content_rating' columns used for merging
merged_data = merged_data.drop(columns=['movie_title', 'content_rating'])

# If you want to save the result to a new CSV file
merged_data.to_csv('imdb_certificate_filled_1.csv', index=False)
```
