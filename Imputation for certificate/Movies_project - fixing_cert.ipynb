```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Step 1: Prepare the data
# Filter rows with missing 'Certificate' values
train_data = imdb_1000_cert_df[imdb_1000_cert_df['Certificate'].notna()]
test_data = imdb_1000_cert_df[imdb_1000_cert_df['Certificate'].isna()]

# Step 2: Clean and convert 'Runtime' column to numeric
# Remove 'min' and convert to integer

# Convert the 'Runtime' column to string to avoid non-string issues
train_data['Runtime'] = train_data['Runtime'].astype(str).str.replace(' min', '').astype(float)
test_data['Runtime'] = test_data['Runtime'].astype(str).str.replace(' min', '').astype(float)


# Step 3: Encode categorical features for training (e.g., Genre, Director)
categorical_cols = ['Genre_1', 'Director', 'Star1', 'Star2', 'Star3', 'Star4'] 
label_encoders = {}

# Handle unseen categories by adding 'Unknown' during the encoding of test data
for col in categorical_cols:
    le = LabelEncoder()
    
    # Fit on train data
    train_data[col] = le.fit_transform(train_data[col].astype(str))
    
    # Check if the test data has unseen categories
    test_data[col] = test_data[col].apply(lambda x: x if x in le.classes_ else 'Unknown')
    
    # Add 'Unknown' to classes if it's not already there
    le.classes_ = np.append(le.classes_, 'Unknown')
    
    # Transform the test data
    test_data[col] = le.transform(test_data[col].astype(str))
    
    label_encoders[col] = le  # Save the encoders for later use if needed

# Step 4: Select relevant features and target variable
X_train = train_data[['IMDB_Rating', 'Runtime', 'Genre_1', 'No_of_Votes']]  # Add other relevant features
y_train = train_data['Certificate']

X_test = test_data[['IMDB_Rating', 'Runtime', 'Genre_1', 'No_of_Votes']]  # Same features for test data

# Step 5: Train Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# Step 6: Predict missing 'Certificate' values
test_data['Certificate'] = rf.predict(X_test)

# Step 7: Reinsert the predicted values back into the original dataset
imdb_1000_cert_df.loc[imdb_1000_cert_df['Certificate'].isna(), 'Certificate'] = test_data['Certificate']

# Now the dataset has imputed 'Certificate' values

# Save the modified dataset with imputed 'Certificate' values to a new CSV file
output_file_path = 'imdb_1000_genre_F_fillttered_imputed.csv'
imdb_1000_cert_df.to_csv(output_file_path, index=False)

#print(f"Dataset saved to {output_file_path}")
```
