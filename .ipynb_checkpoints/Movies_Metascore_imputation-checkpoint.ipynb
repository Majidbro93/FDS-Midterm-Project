```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the imbd datasets for different variables
movies_original_path = 'imdb_top_1000.csv'
# read the CVS file of the movies
imdb_1000_org_df = pd.read_csv(movies_original_path)
```

```python
# This line is for finding out the missing values in the dataset
missing_values = imdb_1000_org_df.isna().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot heatmap to visualize missing values
plt.figure(figsize=(12, 8))
sns.heatmap(imdb_1000_org_df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in IMDb Top 1000 Movies Dataset')
plt.show()
```

```python
# Ensure 'Gross' and 'Runtime' are strings before replacing characters
imdb_1000_org_df['Gross'] = imdb_1000_org_df['Gross'].astype(str).str.replace(',', '').astype(float)
imdb_1000_org_df['Runtime'] = imdb_1000_org_df['Runtime'].astype(str).str.replace(' min', '').astype(float)

# Select relevant columns for correlation
selected_columns = ['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Runtime', 'Gross']

# Compute the correlation matrix
correlation_matrix = imdb_1000_org_df[selected_columns].corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of the numerical variables of the original dataset")
plt.show()
```

```python
# Splitting the 'Genre' column into multiple columns based on ','
genre_split = imdb_1000_org_df['Genre'].str.split(',', expand=True)

# Rename the new genre columns
genre_split.columns = [f'Genre_{i+1}' for i in range(genre_split.shape[1])]

# Concatenate the original dataset with the new genre columns
imdb_1000_gensp_df = pd.concat([imdb_1000_org_df, genre_split], axis=1)

# Display the first few rows of the modified dataset
imdb_1000_gensp_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot heatmap to visualize missing values
plt.figure(figsize=(12, 8))
sns.heatmap(imdb_1000_gensp_df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values of the gensep dataset')
plt.show()
```

```python
from sklearn.impute import KNNImputer
import pandas as pd

# Selecting relevant numerical features
features = ['IMDB_Rating', 'No_of_Votes', 'Meta_score']  # Features including 'Meta_score'
df_knn = imdb_1000_gensp_df[features].copy()

# Applying KNN Imputation
imputer = KNNImputer(n_neighbors=10)
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df_knn), columns=features)

# Replacing the original 'Gross' column with the imputed one
#imdb_1000_gensp_df['Meta_score'] = df_knn_imputed['Meta_score']

imdb_1000_KNNMeta_df = imdb_1000_gensp_df.copy()

imdb_1000_KNNMeta_df['Meta_score'] = df_knn_imputed['Meta_score']
```

```python
import pandas as pd

imdb_1000_MedMeta_df = imdb_1000_gensp_df.copy()
# Fill missing Meta_score values with the median Meta_score within each genre in the copied dataset
imdb_1000_MedMeta_df['Meta_score'] = imdb_1000_MedMeta_df.groupby('IMDB_Rating')['Meta_score'].transform(lambda x: x.fillna(x.median()))
```

```python
# Plot heatmap to visualize missing values
plt.figure(figsize=(12, 8))
sns.heatmap(imdb_1000_gensp_df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in IMDb Top 1000 Movies Dataset')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'Gross' and 'Runtime' are strings before replacing characters for the original dataset
imdb_1000_org_df['Gross'] = imdb_1000_org_df['Gross'].astype(str).str.replace(',', '').astype(float)
imdb_1000_org_df['Runtime'] = imdb_1000_org_df['Runtime'].astype(str).str.replace(' min', '').astype(float)

# Ensure 'Gross' and 'Runtime' are strings before replacing characters for the imputed dataset
imdb_1000_gensp_df['Gross'] = imdb_1000_gensp_df['Gross'].astype(str).str.replace(',', '').astype(float)
imdb_1000_gensp_df['Runtime'] = imdb_1000_gensp_df['Runtime'].astype(str).str.replace(' min', '').astype(float)

# Select relevant columns for correlation
selected_columns = ['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Runtime', 'Gross']

# Compute the correlation matrix for original dataset
correlation_matrix_org = imdb_1000_org_df[selected_columns].corr()

# Compute the correlation matrix for imputed dataset
correlation_matrix_gensp = imdb_1000_gensp_df[selected_columns].corr()

# Create a figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the correlation matrix for original dataset
sns.heatmap(correlation_matrix_org, annot=True, cmap="coolwarm", vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title("Correlation Matrix: Original Dataset")

# Plot the correlation matrix for imputed dataset
sns.heatmap(correlation_matrix_gensp, annot=True, cmap="coolwarm", vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title("Correlation Matrix: Imputed Dataset")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Loading the imbd dataset
movies_mapped_file = 'imdb_mapped_certificate.csv'
# read the CVS file of the movies
imdb_1000_mapped_df = pd.read_csv(movies_mapped_file)
```

```python
imdb_1000_gensp_df['Certificate'] =  imdb_1000_mapped_df['Certificate']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Step 1: Prepare the data
# Filter rows with missing 'Certificate' values
train_data = imdb_1000_gensp_df[imdb_1000_gensp_df['Certificate'].notna()]
test_data = imdb_1000_gensp_df[imdb_1000_gensp_df['Certificate'].isna()]

# Step 2: Clean and convert 'Runtime' column to numeric
# Remove 'min' and convert to integer

# Convert the 'Runtime' column to string to avoid non-string issues
train_data['Runtime'] = train_data['Runtime'].astype(str).str.replace(' min', '').astype(float)
test_data['Runtime'] = test_data['Runtime'].astype(str).str.replace(' min', '').astype(float)


# Step 3: Encode categorical features for training (e.g., Genre, Director)
categorical_cols = ['Genre_1', 'Director', 'Star1', 'Star2', 'Star3', 'Star4']  # Select relevant categorical columns
label_encoders = {}

# Handle unseen categories by adding 'Unknown' during the encoding of test data
for col in categorical_cols:
    le = LabelEncoder()
    
    # Fit on train data
    train_data[col] = le.fit_transform(train_data[col].astype(str))
    
    # Check if the test data has unseen categories
    test_data[col] = test_data[col].apply(lambda x: x if x in le.classes_ else 'Unknown')
    
    # Add 'Unknown' to classes if it's not already there
    le.classes_ = np.append(le.classes_, 'Unknown')
    
    # Transform the test data
    test_data[col] = le.transform(test_data[col].astype(str))
    
    label_encoders[col] = le  # Save the encoders for later use if needed

# Step 4: Select relevant features and target variable
X_train = train_data[['IMDB_Rating', 'Runtime', 'Genre_1', 'No_of_Votes']]  # Add other relevant features
y_train = train_data['Certificate']

X_test = test_data[['IMDB_Rating', 'Runtime', 'Genre_1', 'No_of_Votes']]  # Same features for test data

# Step 5: Train Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# Step 6: Predict missing 'Certificate' values
test_data['Certificate'] = rf.predict(X_test)

# Step 7: Reinsert the predicted values back into the original dataset
imdb_1000_gensp_df.loc[imdb_1000_gensp_df['Certificate'].isna(), 'Certificate'] = test_data['Certificate']

# Now the dataset has imputed 'Certificate' values

# Save the modified dataset with imputed 'Certificate' values to a new CSV file
#output_file_path = 'imdb_1000_imputed_certificate.csv'
#imdb_1000_mapped_df.to_csv(output_file_path, index=False)

#print(f"Dataset saved to {output_file_path}")

```
